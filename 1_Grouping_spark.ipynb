```python
# prompt: male ratio per occuption

import pandas as pd
import os

!pip install pandas

# Create a dummy student.csv for testing purposes
data = {
    'user_id|age|gender|occupation|zip_code': [
        '1|24|M|technician|85711',
        '2|53|F|other|94043',
        '3|23|M|writer|32067',
        '4|24|M|technician|43537',
        '5|33|F|other|15213',
        '6|42|M|executive|98101',
        '7|57|M|administrator|91344',
        '8|36|M|administrator|05201',
        '9|29|M|student|01002',
        '10|31|M|programmer|85202',
        '11|31|F|other|01002',
        '12|25|M|writer|01002',
        '13|47|M|marketing|01002',
        '14|30|F|marketing|01002'
    ]
}

df = pd.DataFrame(data)
df.to_csv('student.csv', index=False)

# Load the data, handling potential errors
file_path = 'student.csv'
if os.path.exists(file_path) and os.path.getsize(file_path) > 0:
  try:
    users = pd.read_csv(file_path, sep=',')
    users = pd.DataFrame(users['user_id|age|gender|occupation|zip_code'].str.split('|').tolist())
    users.columns = ['user_id', 'age', 'gender', 'occupation', 'zip_code']

    # Calculate the number of males per occupation
    male_counts = users[users['gender'] == 'M'].groupby('occupation')['gender'].count()

    # Calculate the total number of people per occupation
    total_counts = users.groupby('occupation')['gender'].count()

    #Calculate male ratio per occupation
    male_ratio = male_counts / total_counts
    print('male ratio per occupation')
    print(male_ratio)
  except FileNotFoundError:
    print("Error: 'student.csv' not found. Please ensure the file exists in the current directory.")
  except KeyError:
    print("Error: 'user_id|age|gender|occupation|zip_code' column not found. Please ensure the column name is correct")
  except pd.errors.EmptyDataError:
      print("Error: 'student.csv' is empty.")
  except Exception as e:
      print(f"An unexpected error occurred: {e}")
else:
  print(f"Error: File '{file_path}' does not exist or is empty.")
```

```python
# prompt: For each occupation, calculate the minimum and maximum ages

!pip install pandas
import pandas as pd
import os

# Create a dummy student.csv for testing purposes
data = {
    'user_id|age|gender|occupation|zip_code': [
        '1|24|M|technician|85711',
        '2|53|F|other|94043',
        '3|23|M|writer|32067',
        '4|24|M|technician|43537',
        '5|33|F|other|15213',
        '6|42|M|executive|98101',
        '7|57|M|administrator|91344',
        '8|36|M|administrator|05201',
        '9|29|M|student|01002',
        '10|31|M|programmer|85202',
        '11|31|F|other|01002',
        '12|25|M|writer|01002',
        '13|47|M|marketing|01002',
        '14|30|F|marketing|01002'
    ]
}

df = pd.DataFrame(data)
df.to_csv('student.csv', index=False)

# Load the data, handling potential errors
file_path = 'student.csv'
if os.path.exists(file_path) and os.path.getsize(file_path) > 0:
  try:
    users = pd.read_csv(file_path, sep=',')
    users = pd.DataFrame(users['user_id|age|gender|occupation|zip_code'].str.split('|').tolist())
    users.columns = ['user_id', 'age', 'gender', 'occupation', 'zip_code']

    # Convert 'age' to numeric
    users['age'] = pd.to_numeric(users['age'], errors='coerce')

    # Group by occupation and calculate min and max ages
    min_max_ages = users.groupby('occupation')['age'].agg(['min', 'max'])
    print("Minimum and maximum ages per occupation:")
    print(min_max_ages)

  except FileNotFoundError:
    print("Error: 'student.csv' not found. Please ensure the file exists in the current directory.")
  except KeyError:
    print("Error: 'user_id|age|gender|occupation|zip_code' column not found. Please ensure the column name is correct")
  except pd.errors.EmptyDataError:
    print("Error: 'student.csv' is empty.")
  except Exception as e:
    print(f"An unexpected error occurred: {e}")

else:
  print(f"Error: File '{file_path}' does not exist or is empty.")
```
